```
from keras.datasets import mnist
import matplotlib.pyplot as plt
import numpy as np
```

```
(train_images, train_labels), (test_images, test_labels) = mnist.load_data()
```

```
def printImages(mean):
  plt.figure(figsize=(4,5))
  plt.imshow(mean)
  plt.show()
```

```
def filter(labels,n):
  return np.where(labels == n)[0]
```

```
imageRange = 10
axis = 0
for i in range(imageRange):
  mean = np.mean(train_images[filter(train_labels, i)], axis = axis)
  
  printImages(mean)
```
