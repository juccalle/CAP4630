```
from mpl_toolkits import mplot3d
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
```

```
#Function One
def f(x,y):
  return x ** 2 - y ** 2

x = np.linspace(-6, 6, 30)
y = np.linspace(-6, 6, 30)

X,Y = np.meshgrid(x,y)
Z = f(X,Y)

shape = plt.figure()

onex = plt.axes(projection='3d')
onex.contour3D(X,Y,Z,25,cmap='binary')
onex.set_xlabel('x')
onex.set_ylabel('y')
onex.set_zlabel('z')
onex.scatter([0],[0],[0], color='b', s=20)
```

```
x = np.linspace(-5,5,50)
y = np.linspace(-5,5,50)

X,Y = np.meshgrid(x,y)
Z = f(X,Y)
plt.contourf(X,Y,Z,20,cmap='Blues')
plt.colorbar();
plt.plot(0,0,'bo')
```

```
#Function Two
def g(x,y):
  return (x ** 4-4*x**2) - (y ** 4-4*y**2)

x = np.linspace(-1, 1, 30)
y = np.linspace(-1, 1, 30)

X,Y = np.meshgrid(x,y)
Z = g(X,Y)

shape = plt.figure()

onex = plt.axes(projection='3d')
onex.contour3D(X,Y,Z,60,cmap='binary')
onex.set_xlabel('x')
onex.set_ylabel('y')
onex.set_zlabel('z')
onex.scatter([0],[0],[0], color='b', s=30)
onex.scatter([0],[1],[3], color='g', s=30)
onex.scatter([0],[-1],[1], color='g', s=30)
onex.scatter([1],[0],[-3], color='r', s=30)
onex.scatter([-1],[0],[-3], color='r', s=30)
```

```
x = np.linspace(-2,2,50)
y = np.linspace(-2,2,50)

X,Y = np.meshgrid(x,y)
Z = g(X,Y)
plt.contourf(X,Y,Z,20,cmap='Blues')
plt.colorbar();
plt.plot(0,0,'bo')
plt.plot(0,1,'go')
plt.plot(0,-1, 'go')
plt.plot(1,0,'ro')
plt.plot(-1,0,'ro')
```

```
#Function Three
def h(x,y):
  return -2*x ** 2 - 4*y ** 2

x = np.linspace(-6, 6, 30)
y = np.linspace(-6, 6, 30)

X,Y = np.meshgrid(x,y)
Z = h(X,Y)

shape = plt.figure()

onex = plt.axes(projection='3d')
onex.contour3D(X,Y,Z,15,cmap='binary')
onex.set_xlabel('x')
onex.set_ylabel('y')
onex.set_zlabel('z')
onex.scatter([0],[0],[0], color='b', s=20)
```
